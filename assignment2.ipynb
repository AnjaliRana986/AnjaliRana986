```python
#write a code in python for the 8 puzzle problem by taking the initial and final state
import copy

class eightPuzzle:
    def __init__(self,start_pos,goal_pos):
        self.mygrid = start_pos
        self.mygoal = goal_pos
        self.myindex = self.getIndex()
        self.prevState = None
        self.displayState()
    
    def getIndex(self):
        for i in range(9):
            if(self.mygrid[i]==0):
                return i
    
    def displayState(self):
        for i in range(0,9,3):
            print(self.mygrid[i],self.mygrid[i+1],self.mygrid[i+2])
        print()
    
    def up(self):
        if(self.myindex == 6 or self.myindex == 7 or self.myindex == 8):
            # print('Incorrect Move!!')
            return False
        else:
            self.prevState = copy.deepcopy(self)
            temp = self.mygrid[self.myindex + 3]
            self.mygrid[self.myindex + 3] = 0
            self.mygrid[self.myindex] = temp
            self.myindex = self.myindex + 3
            self.displayState()
            return True
    def down(self):
        if(self.myindex == 0 or self.myindex == 1 or self.myindex == 2):
            # print('Incorrect Move!!')
            return False
        else:
            self.prevState = copy.deepcopy(self)
            temp = self.mygrid[self.myindex - 3]
            self.mygrid[self.myindex - 3] = 0
            self.mygrid[self.myindex] = temp
            self.myindex = self.myindex - 3
            self.displayState()
            return True
    def left(self):
        if(self.myindex == 2 or self.myindex == 5 or self.myindex == 8):
            # print('Incorrect Move!!')
            return False
        else:
            self.prevState = copy.deepcopy(self)
            temp = self.mygrid[self.myindex + 1]
            self.mygrid[self.myindex + 1] = 0
            self.mygrid[self.myindex] = temp
            self.myindex = self.myindex + 1
            self.displayState()
            return True
    def right(self):
        if(self.myindex == 0 or self.myindex == 3 or self.myindex == 6):
            # print('Incorrect Move!!')
            return False
        else:
            self.prevState = copy.deepcopy(self)
            temp = self.mygrid[self.myindex - 1]
            self.mygrid[self.myindex - 1] = 0
            self.mygrid[self.myindex] = temp
            self.myindex = self.myindex - 1
            self.displayState()
            return True
    
    def goalReached(self):
        return self.mygrid == self.mygoal

    def __eq__(self,other):
        return self.mygrid == other.mygrid
    
    def nextPossibleState(self):
        stateList=[]

        upState=copy.deepcopy(self)
        if upState.up():
            stateList.append(upState)

        downState=copy.deepcopy(self)
        if downState.down():
            stateList.append(downState)

        leftState=copy.deepcopy(self)
        if leftState.left():
            stateList.append(leftState)

        rightState=copy.deepcopy(self)
        if rightState.right():
            stateList.append(rightState)

        return stateList
            

def BFS(startState):
    open = [startState]
    closed = []
    while(open):
        this_state = open.pop(0)
        if this_state.goalReached():
            print('Goal reached')
            closed.append(this_state)
            this_state.displayState()
            break
        else:
            closed.append(this_state)
            next_state = this_state.nextPossibleState()
            for each_state in next_state:
                if each_state not in open and each_state not in closed:
                    open.append(each_state)
    

# puzzle = eightPuzzle([7,6,5,1,0,2,3,4,8],[0,1,2,3,4,5,6,7,8])
puzzle = eightPuzzle([1,2,3,8,0,4,7,6,5],[2,8,1,0,4,3,7,6,5])
# puzzle = eightPuzzle([7,6,5,1,0,2,3,4,8],[7,6,0,1,4,5,3,8,2])
BFS(puzzle)
```

```python
# Travelling Sales man problem
graph=[[0, 10, 15, 20],
       [10, 0, 35, 25],
       [15, 35, 0, 30],
       [20, 25, 30, 0]]
from itertools import permutations
l = list(permutations(range(1, 4+1))) 
print(l)
min=999  
for x in l:
    if x[0]==1:        
        print ("Route:", x) 
        sum=0    
        for j in range(0,len(x)):
            if (j==(len(x)-1)):
                sum=sum+graph[x[j]-1][x[0]-1]
                print("Cost:", sum)
            else:
                sum=sum+graph[x[j]-1][x[j+1]-1]                               
        if sum<min:
            min=sum
            temp=x

print("\nBest Route:", temp)
print("Cost:", min)
```
